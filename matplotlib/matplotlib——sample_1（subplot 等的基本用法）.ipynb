```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

p = np.arange(0.1, 1, 0.2)
u = np.arange(10, 60, 10)
x = np.arange(0, 0.5, 0.01)
plt.figure(figsize = (15, 10))

for i in range(5):
    pi = p[i]
    ui = u[i]
    ED_FIFO = pi / (ui - pi * ui) + x
    ED_PS = x / (1 - pi)
    plt.subplot(2, 3, i + 1)
    plt.title('p=%.1f u=%d' % (pi,ui))
    plt.plot(x, ED_FIFO, label='M/M/1-FIFO')
    plt.plot(x, ED_PS, color='red', linewidth=1.0, linestyle='--', label='M/M/1-PS')
    plt.legend(loc='upper right')
    plt.xlabel('Requirement Time x')
    plt.ylabel('Mean Delay')

plt.show()
```
